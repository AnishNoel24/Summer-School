```python
import matplotlib.pyplot as plt

file_sizes = [200, 400, 600, 800, 1000]  # File sizes in MB
languages = ['C', 'C++', 'Java', 'R', 'Python']

execution_times = [
    [12, 15, 18, 20, 25],    # Time taken for 200MB
    [20, 25, 30, 35, 40],    # Time taken for 400MB
    [34, 36, 40, 45, 53],    # Time taken for 600MB
    [45, 50, 55, 60, 75],    # Time taken for 800MB
    [55, 60, 70, 80, 100]    # Time taken for 1000MB
]

# Plotting the graph
plt.figure(figsize=(10, 6))
for i in range(len(languages)):
    plt.plot(file_sizes, execution_times[i], marker='o', label=languages[i])

plt.xlabel('File Size (MB)')
plt.ylabel('Time Taken (sec)')
plt.title('Performance Comparison: Text File Conversion to Uppercase')
plt.legend()
plt.grid(True)
plt.show()
```
