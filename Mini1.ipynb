1. Problem Statement:
Find all the subsets from a set of numbers whose sum is zero.
Constraint: Subset size must be 5
Set={-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}

```python
from itertools import combinations

def find_subsets(n):
    subsets = []
    for subset in combinations(n, 5):
        if sum(subset) == 0:
            subsets.append(subset)
    return subsets


n = [-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4]
zero_sum_subsets = find_subsets(n)
for subset in zero_sum_subsets:
    print(subset)
print("TOTAL NUMBER OF SUBSETS WITH ZERO SUM=",len(zero_sum_subsets))

```

 Problem Statement:
Find all the subsets from a set of numbers whose sum is zero.
Constraint: Subset size must be 3 to 6 only
Set={-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}

```python
from itertools import combinations

def find_subsets(n):
    subsets = []

    # Generate all combinations of size 3 to 6
    for r in range(3, 7):
        for subset in combinations(n, r):
            # Check if the sum of the subset is zero
            if sum(subset) == 0:
                subsets.append(subset)

    return subsets

# Set of numbers
n = [-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4]

# Find subsets with zero sum and size 3 to 6
zero_sum_subsets = find_subsets(n)

# Print the subsets
for subset in zero_sum_subsets:
    print(subset)
print("TOTAL NUMBER OF SUBSETS WITH ZERO SUM=",len(zero_sum_subsets))
```
